```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib as mpl

from matplotlib import rc
from matplotlib import font_manager

import warnings

# 한글 글꼴
font_path = 'C:/Users/YJ/AppData/Local/Microsoft/Windows/Fonts/NotoSansKR-Black.ttf'
font_prop = font_manager.FontProperties(fname= font_path)
rc('font', family = font_prop.get_name())

# 음수 깨짐 방지
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = sns.load_dataset('tips')
```

```python
fig, ax = plt.subplots(2,2)
fig.set_size_inches(16,8)

sns.scatterplot( x = df['total_bill']
                , y = df['tip']
                , ax = ax[0][0])

ax[0][0].set(title = '산점도')

sns.barplot(x = df['size']
            , y = df['tip']
            , ax = ax[0][1]
)

ax[0][1].set(title = '막대차트')

sns.regplot(x = df['total_bill']
            , y = df['tip']
            , ax = ax[1][0]
            )

ax[1][0].set(title = 'regplot')

sns.violinplot(x = df['size']
               , y = df['tip']
               , ax = ax[1][1]
               )

ax[1][1].set(title = '바이올린차트')

plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(2,2, sharex = True)

fig.set_size_inches(16,8)

sns.pointplot(x = df['size']
              , y = df['tip']
              , ax = ax[0][0]
              )

ax[0][0].set(title='pointplot')

sns.barplot(x = df['size']
            , y = df['tip']
            , ax = ax[0][1]
            )

ax[0][1].set(title='barplot')

sns.boxplot(x = df['size']
            , y = df['tip']
            , ax = ax[1][0])

ax[1][0].set(title = 'boxplot')

sns.violinplot(x = df['size']
               , y = df['tip']
               , ax = ax[1][1]
               )

ax[1][1].set(title='violinplot')
```

```python
fig, _ = plt.subplots()
fig.set_size_inches(10,4)

sns.boxplot(x = df['size']
            , y = df['tip']
            )

plt.title('BoxPlot')
plt.grid()

```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(12,4)

sns.boxplot(x = df['smoker']
            , y = df['tip']
            , ax = ax1
            )

sns.boxplot(x = df['sex']
            , y = df['tip']
            , ax = ax2
            )

fig, ax3 = plt.subplots(1,1)
fig.set_size_inches(12,4)

sns.scatterplot(x = df['total_bill']
                , y = df['tip']
                , hue = df['size']
                , ax = ax3
                )
```
