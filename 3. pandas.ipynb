- 잡단의 데이터가 쏠리는 현상을 방지하고, 모집단에서 집단 간의 비율을 고려하여 샘플링을 진행

```python
# 값 치환
tmp_line = sorted(sample_df['노선명'].unique())[:4]
tmp_eng_line = ['line' + str(i) for i in range(1,5)]

sample_df['노선명'].replace(tmp_line, tmp_eng_line, inplace = True)

sample_df
```

```python
def getDate(data):
    data = str(data)
    year = data[0:4]
    month = data[4:6]
    day = data[6:8]
    return year + '-' + month + '-' + day
```

```python
d = 20210201
getDate(d)
```

```python
data['사용일자1'] = data['사용일자'].apply(getDate)
data
```

#### 비율별 결측치 처리방법
- 10% 미만 : 로우 삭제 또는 치환
- 10 ~ 15% : 모델 기반 처리
- 50% 이상 : 컬럼 삭제

```python
# 성적 랜덤으로 생성
# numpy로 결측치 생성 가능
import numpy as np

np.random.seed(41)

rand_list1 = [np.random.randint(10, 95) // 10 * 10 for i in range(10)]
min_value_index = rand_list1.index(min(rand_list1))


for i in range(0,10):
    if i == min_value_index:
        rand_list1[i] = np.nan

np.random.seed(42)

rand_list2 = [np.random.randint(10, 95) // 10 * 10 for i in range(10)]
min_value_index = rand_list2.index(min(rand_list2))


for i in range(0,10):
    if i == min_value_index:
        rand_list2[i] = np.nan

np.random.seed(43)

rand_list3 = [np.random.randint(10, 95) // 10 * 10 for i in range(10)]
min_value_index = rand_list3.index(min(rand_list3))


for i in range(0,10):
    if i == min_value_index:
        rand_list3[i] = np.nan

rand_list1, rand_list2, rand_list3
```

```python
# 랜덤으로 생성된 리스트 부여

data = {
    '학번' : range(2000, 2010)
    , '국어성적' : rand_list1
    , '영어성적' : rand_list2
    , '수학성적' : rand_list3
}

df = pd.DataFrame(data = data)
df
```

```python
# 결측치 확인
df.isna().sum()
```

```python
# 결측치 0으로 채우기
df.fillna(0)
```
