```python
import numpy as np
import matplotlib.pyplot as plt

# 비용 함수 정의
def cost_function(w):
    return (w - 2) ** 2 + 1  # 예시로 간단한 비용 함수 사용

# x축 범위 설정
w = np.linspace(-2, 6, 100)
cost = cost_function(w)

# 비용 함수 그래프
plt.figure(figsize=(8, 6))
plt.plot(w, cost, label='Cost Function J(w)', color='blue')
plt.xlabel('w (Weight)')
plt.ylabel('J(w) (Cost)')
plt.title('Cost Function Graph')
plt.legend()
plt.grid()
plt.show()
```
