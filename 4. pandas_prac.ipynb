```python
import pandas as pd
import numpy as np
from pandas_datareader import data
import yfinance as yfin
#yfin.pdr_override() #0.1.96 버전만 있음
```

```python
# 삼성 데이터
df = yfin.download('005930.KS', start = '2020-01-01', end = '2024-12-24')
df
```

```python
df = df.droplevel('Ticker',axis = 1).reset_index()
df.columns.name = None
df
```

```python
# 주요 통계 지표 확인
df.describe()
```

```python
# 종가(close) 기준 최대값
df['Close'].max()
```

```python
# 종가 기준 최소값
df['Close'].min()
```

```python
# 종가 기준 평균값
df['Close'].mean()
```

```python
# 총 데이터 개수
a, _ = df.shape
print(f'총 row 개수 : {a}')

print(f'총 row 개수 : {len(df)}')
```

```python
# 특정 컬럼만 선택(close , higt/low 2개의 컬럼 선택)
df[['Close', 'High', 'Low']]
```

```python
# 상위 3개
df[:3]
df.head(3)
```

```python
# 시작가보다 종가가 높더 날
df[df['Open'] < df['Close']]
```

```python
# 종가 기준 가장 주가가 높았던 날
df[df['Close'] == max(df['Close'])]
```

```python
# 종가 기준 주가가 8만원 이상이던 날짜
df[df['Close'] >= 80000]['Date'].unique()
```
