```python
# date > 날짜, High > 고가 Low > 저가
df.rename(columns = {'Date' : '날짜'
                     , 'High' : '고가'
                     , 'Low' : '저가'}
        , inplace= True)


df
```

```python
# 날짜 컬럼 삭제
df.drop(columns = '날짜')
```

```python
# 종가 시가의 차이 계산 컬럼
df['diff'] = df['Close'] - df['Open']
df
```

```python
# Date 필드에 저장된 값을 파싱하여, Year, Month, Day 필드에 저장
df['Year'] = [i.year for i in df['날짜']]
df['Month'] = [i.month for i in df['날짜']]
df['Day'] = [i.day for i in df['날짜']]

df
```

#### `apply` vs `applymap` in Pandas
| **특징**           | **`apply`**                              | **`applymap`**                          |
|---------------------|------------------------------------------|------------------------------------------|
| **적용 범위**       | 행(row) 또는 열(column) 전체             | 데이터프레임의 각 개별 요소(element)     |
| **적용 대상**       | 행, 열, 또는 시리즈 단위                 | 데이터프레임의 모든 개별 값              |
| **함수 작동 방식**  | 열 또는 행 단위로 작동                  | 각 값에 대해 작동                        |
| **결과 형태**       | 시리즈(Series) 또는 데이터프레임(DataFrame) | 데이터프레임(DataFrame)                 |


```python
# 모든 데이터를 소수점 둘쨰자리까지 표시
df.applymap(lambda x : '%.2f' % x if isinstance(x, (int, float)) else x)
```

```python
# 천 단위 구분 기호
df['Volume'].apply(lambda x : "{:,}".format(x))
```

```python
# month로 groupby해서 평균 구하기
df['Volume'].groupby(df['Month']).mean()
```
