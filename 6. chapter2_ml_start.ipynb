### 모델을 위한 데이터 수집

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris_data = iris.data

# 4값
iris_data # 4개의 특성(features)
```

```python
# y값
iris_label = iris.target
iris_label, iris.target_names
```

```python
iris.feature_names
```

```python
import pandas as pd

iris_df = pd.DataFrame(iris_data, columns = iris.feature_names)

iris_df
```

```python
iris_df.head()
```

### 데이터 분할

```python
from sklearn.model_selection import train_test_split

# X : features
# y : label(실제값)

X_train, X_test, y_train, y_test = \
train_test_split(iris_data, iris_label, test_size = 0.2, random_state=11) # random_state 아무 숫자 지정해도 됨(seed같은 느낌)
```

#### 모델 생성

```python
from sklearn.tree import DecisionTreeClassifier

# 깡통 생성
dt_clf = DecisionTreeClassifier(random_state=11)
dt_clf

#df_clf는 예측기라고 부름
```

### 모델 학습

```python
dt_clf.fit(X_train, y_train)
```

### 예측

```python
# y_test를 return 함 > 예측 라벨값
pred = dt_clf.predict(X_test)
pred
```

```python
# 정답 y_test, pred의 예측값을 비교 > 정답률(점수)
from sklearn.metrics import accuracy_score

accuracy_score(y_test,pred)
```

### 훈련/테스트 데이터 분할 >> "홀드 아웃"

### 교차검증 CrossValidation

```python
from sklearn.model_selection import KFold
import numpy as np

kfold = KFold(n_splits = 5)
df_clf = DecisionTreeClassifier(random_state = 156)
df_clf
```

```python
features = iris.data
label = iris.target

cv_accuracy = []

fold_data = []

for train_index, test_index in kfold.split(features): # 5개로 쪼갬
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    df_clf.fit(X_train, y_train)
    pred = df_clf.predict(X_test)

    accuracy = np.round(accuracy_score(y_test, pred), 5)
    cv_accuracy.append(accuracy)
    fold_data.append((X_train, y_train, X_test, y_test))

    print(X_train.shape[0], X_test.shape[0])


print(f'''
예측 : {cv_accuracy}
예측 평균 : {np.mean(cv_accuracy)}
      ''')
```

```python
import matplotlib.pyplot as plt

# 교차 검증 데이터의 분포 시각화
fig,axes = plt.subplots(1,5, figsize= (20,5), sharex=True, sharey=True)

for i, (X_train, y_train, X_test, y_test) in enumerate(fold_data):
    axes[i].scatter(X_train[:,0], X_train[:, 1], c=y_train, cmap = 'coolwarm', alpha = 0.6, label = 'Train')
    axes[i].scatter(X_test[:,0], X_test[:,1], c=y_test, cmap = 'coolwarm', edgecolor = 'k', marker='s', label = 'Test')
    axes[i].set_title(f'Fold {i + 1}')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

### Stratified k-fold Cross Validation
데이터의 분포

```python
import pandas as pd
iris = load_iris()

iris_df = pd.DataFrame(data = iris.data, columns = iris.feature_names)

iris_df['label'] = iris.target
iris_df['label'].value_counts()

```

```python
# k-fold = 3
#직접 해보기
```

### Stratified Split 분포 확인

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=3)

for train_index, test_index in skf.split(iris_df, iris_df['label']):
    label_train = iris_df['label'].iloc[train_index]
    label_test = iris_df['label'].iloc[test_index]

    print(f'''{'*'*50}
train 분포 : {label_train.value_counts()}

test 분포 : {label_test.value_counts()}
{'*'*50}
''')

```

### 교차 검증 모델 생성

```python
dt_clf = DecisionTreeClassifier(random_state=156)
dt_clf
```

```python
skfold = StratifiedKFold(n_splits = 3)

cv_accuracy = []
for train_index, test_index in skfold.split(features, label):

    # 데이터 분할
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    # print(X_train, X_test, y_train, y_test)

    # 모델 학습(n = 3)
    dt_clf.fit(X = X_train, y = y_train)

    # 예측
    pred = dt_clf.predict(X_test)

    # 정확도 검증
    cv_accuracy.append(accuracy_score(y_test, y_pred = pred))

cv_accuracy, np.mean(cv_accuracy)
```

### cross_val_score()

```python
dt_clf1 = DecisionTreeClassifier(random_state=156)
dt_clf1
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(dt_clf1
                        , iris.data
                        , label
                        , scoring = 'accuracy'
                        , cv = 3 # stratified
                        )

scores, np.mean(scores)
# 위와 같은 효과임
```

### GridSearchCV - 하이퍼 파라미터 튜닝 + 교차검증

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size = 0.2, random_state=121)

dt_clf2 = DecisionTreeClassifier(random_state=121)

# parameter grid
parameters = {
            'max_depth' : [1,2,3]
            , 'min_samples_split' : [2,3]
            }
```

```python
from sklearn.model_selection import GridSearchCV

# model, model의 하이퍼 파라미터의 목록을 지정, 교차검증 파라미터 부여하고 한번에 모델을 만든다.
grid_dtree = GridSearchCV(dt_clf2
                        , param_grid=parameters
                        , cv = 3
                        , refit = True # default
                        )
grid_dtree
```

```python
grid_dtree.fit(X_train, y_train)

# 학습 결과 : best_estimator을 찾아줌(최적의 모델 파라미터를 찾는다)
```

```python
# 학습 결과
score_df = pd.DataFrame(grid_dtree.cv_results_)
score_df
```

### 최적의 파라미터와 정확도

```python
grid_dtree.best_params_, grid_dtree.best_score_
```

```python
# 최적의 모델
estimator = grid_dtree.best_estimator_

# 예측
pred = estimator.predict(X_test)

# 정확도 평가
accuracy_score(y_test, pred)
```
