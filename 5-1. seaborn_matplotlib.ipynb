#### 회귀선
개별 데이터 분포를 대표할 수 있는 하나의 선인 회귀선(regression)을 표시하는 그래프

```python
plt.figure(figsize = (8,6))

sns.regplot(x = df['total_bill']
            , y = df['tip']
            )

plt.title('회귀선')
```

```python
plt.figure(figsize = (8,6))

sns.regplot(x = df['total_bill']
            , y = df['tip']
            , scatter= False # 회귀선만 봄
            )

plt.title('회귀선만 봄')
```

```python
plt.figure(figsize = (8,6))

sns.regplot(x = df['total_bill']
            , y = df['tip']
            , scatter= False # 회귀선만 봄
            , ci= None # 신뢰구간을 없앰
            )

plt.title('신뢰구간 없앤 회귀선만 봄')
```

```python
plt.figure(figsize = (8,6))

sns.barplot(x = df['size']
            , y = df['tip']
            )

plt.title('막대차트, barplot')
```

```python
plt.figure(figsize = (8,6))

sns.barplot(x = df['size']
            , y = df['tip']
            , ci = None)

plt.title('신뢰구간인 중심선을 없앤 barplot')
```

```python
plt.figure(figsize = (8,6))

sns.countplot(x = df['size'])

plt.title('countplot을 이용한 x축 카테고리별 개수 세기')
```

```python
plt.figure(figsize = (8,6))

sns.pointplot(x = df['size']
              , y= df['tip']
              , hue = df['day'])

plt.title('pointplot을 이용한 데이터 평균 선')
```

```python
plt.figure(figsize = (8,6))

sns.boxplot(x = df['size']
            , y = df['tip']
            , hue = df['day']
            )


plt.title('boxplot 이용한 범주별 통계치')
```

```python
plt.figure(figsize = (8,6))

sns.boxplot(x = df['tip']
            , y= df['size']
            , hue = df['day']
            , orient = 'horizontal')

plt.title('boxplot을 이용한 통계치 확인(수평)')
```

```python
plt.figure(figsize = (8,6))

sns.violinplot(x = df['size']
               , y = df['tip']
               , palette= 'coolwarm'
               )

plt.title('바이올린 차트를 이용한 범주별 분포 확인')
```

```python
plt.figure(figsize= (8,6))

sns.histplot(data = df['total_bill']
             , bins = 5
             , kde = True)


plt.title('histplot을 통한 연속적인 변수 분포 막대')
```

```python
plt.figure(figsize = (8,6))

sns.kdeplot(data = df['total_bill'])

plt.title('kdeplot 활용')
```

```python
plt.figure(figsize = (8,6))

sns.rugplot(data = df['total_bill'])

plt.title('rugplot 활용_바닥에 카페트가 깔리는 모양')
```

```python
df.pivot_table(values = 'tip'
               , index = 'day'
               , columns = 'size'
               , aggfunc= 'count' # 집계함수를 정의할 수 있음(mean, count, sum, median etc)
               )
```

```python
plt.figure(figsize = (8,6))

sns.heatmap(data = df.pivot_table(values = 'tip'
                                , index = 'day'
                                ,columns='size')\
                        .fillna(0)
                                
            , annot= True
            , cmap = 'coolwarm'
            )

plt.title('히트맵')
```

```python
x = [i for i in range(1,13)]
y = [i for i in range(10,130, 10)]

fig = plt.figure(figsize = (10,5))

plt.bar(x
        , y
        , width = 1
        , align = 'edge'
        , color = 'red'
        , edgecolor = 'blue'
        , linewidth = 2
        )
plt.title('matplotlib bar chart')
plt.xlabel('x축은 xlabel')
plt.ylabel('y축은 ylabel')
plt.legend(['cost'])
plt.grid(True, axis = 'y')
plt.show()

```
